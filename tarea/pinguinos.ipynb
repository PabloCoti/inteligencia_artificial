```python
W3 = np.array([[-0.612028000170483,-0.205392985275009,0.236326796364515],
[1.26429338845952,0.298973309266565,0.154285273667061],
[0.209570844845207,-1.33381388798408,-1.01997547619644],
[0.0794451871608174,0.67181973156564,0.31314088150116]])
```

```python
#def f_act(X):
#    activada = np.array([1/(1 + math.exp(-x)) for x in X])
#    return activada

def f_act(X):
    activada = []
    for x in X:
        try:
            r = 1 / (1 + math.exp(-x))
        except OverflowError:
            r = 0
        activada.append(r)
    return np.array(activada)
```

```python
# Carga de datos
#iris = sns.load_dataset('iris')
#iris
penguins = sns.load_dataset('penguins')
penguins
```

```python
xcols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
xcolsp = ['species', 'island','bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']
```

```python
especie = penguins['species'].unique()
especie
```

```python
X = penguins[xcols].copy()
X.insert(0, 'bias', 1)
```

```python
X
```

```python
prediccion = []
for index, fila in X.iterrows():
    capa1 = f_act(fila.dot(W1))
    capa1 = np.insert(capa1, 0, 1)
    capa2 = f_act(capa1.dot(W2))
    capa2 = np.insert(capa2, 0, 1)
    salida = f_act(capa2.dot(W3))
    prediccion.append(especie[np.argmax(salida)])
```

```python
penguins['Prediccion'] = prediccion
penguins.head()
```
